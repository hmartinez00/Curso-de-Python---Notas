```python
def buscar_parrafo(name_file, var_sub_cadena):

    f = open(name_file, 'rb')
    document = Document(f)
    f.close()

    var_parrafos = document.paragraphs
    result = {}

    j = 0
    for i in range(len(var_parrafos)):
        if var_sub_cadena in var_parrafos[i].text:
            result[j] = (i, var_parrafos[i].text)
            j += 1
    
    return result
```
